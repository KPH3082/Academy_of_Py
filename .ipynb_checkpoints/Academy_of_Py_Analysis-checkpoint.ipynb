```python
#Import dependencies
import pandas as pd
import numpy as np

#File paths for csv's

file_path_schools = "PyCitySchools/Resources/schools_complete.csv"
file_path_students = "PyCitySchools/Resources/students_complete.csv"

#Read csv's
schools_data = pd.read_csv(file_path_schools)
students_data = pd.read_csv(file_path_students)

#Seeing what they have in common
print(schools_data.head())
print(students_data.head())
```

## Setting up the dataframes

```python
school_data_complete = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Analyzing the dataframe

```python
#Checking data types that will be good to know later for manipulation
school_data_complete.dtypes
```

```python
#Checking data to look for any NaN's
print (school_data_complete.isnull().sum())
```

```python
#Checking the column names
school_data_complete.columns
```

```python
#Setting a new index for Student ID
school_new_index = school_data_complete.set_index('Student ID')
school_new_index.head()
```

## Getting the numbers

```python
number_of_schools = len(school_new_index['school_name'].unique())
print(number_of_schools)
```

```python
#Checking if the number of students are unique
number_of_students = len(school_new_index['student_name'].unique())
print(number_of_students)
```

```python
#Names are repeated, checking to see if there are duplicate names or different people
#Appears the repeated names are different people in different grades and different schools so going to assume every
#student name is a new person
pd.concat(g for _, g in school_new_index.groupby("student_name") if len(g) > 1)
```

```python
#Calculate the total number of students
tot_students = '{:,}'.format(school_new_index['student_name'].count())

#Calculate the total budget
tot_budget = '${:,.2f}'.format(0)

#Calculate the average math score
avg_math_score = school_new_index['math_score'].mean()

#Calculate the average reading score
avg_reading_score = school_new_index['reading_score'].mean()

#Calculate the overall passing rate (overall average score)
overall_pass_rate = (avg_math_score + avg_reading_score) / 2

#Calculate the percentage of students with a passing math score (70 or greater)
pass_rate_math = (len(school_new_index[school_new_index.math_score >= 70]) / school_new_index.shape[0]) * 100

#Calculate the percentage of students with a passing reading score (70 or greater)
pass_rate_read = (len(school_new_index[school_new_index.reading_score >= 70]) / school_new_index.shape[0])*100



#school_new_index['budget'].sum()
```

```python
district = (number_of_schools, tot_students, tot_budget, 
            avg_math_score, avg_reading_score, pass_rate_math,
            pass_rate_read, overall_pass_rate)
District_Summary = pd.DataFrame([district], columns= ['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math',
            '% Passing Reading','% Overall Passing Rate'])
```

```python
#Finding 'Per Student Budget'
school_data_complete['Per Student Budget'] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete['Per Student Budget'] = school_data_complete['Per Student Budget'].map("${:.2f}".format)

#Passing Rates
school_data_complete['% Passing Math'] = school_data_complete['budget'] / school_data_complete['size']

school_comparison = school_data_complete[['school_name', 'type', 'size', 'budget','Per Student Budget',
                                          'math_score','reading_score']]

school_comparison = school_comparison.groupby(['school_name']).agg({'type':'first', 
                                                'size':'first', 'budget':'first',
                                                'Per Student Budget': 'first',
                                                'math_score': 'mean', 'reading_score':'mean'})
school_comparison.rename(inplace=True, columns={'type':"School Type", 'size':'Total Students',
                                               'budget':'Total School Budget', 
                                                'math_score':'Average Math Score',
                                               'reading_score':'Average Reading Score'})
school_comparison["Total School Budget"] = school_comparison["Total School Budget"].map("${:,.2f}".format)
```
